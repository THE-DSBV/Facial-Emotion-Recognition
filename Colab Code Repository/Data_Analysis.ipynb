```python
# Plot based on FMAE's output on Testing Data
import matplotlib.pyplot as plt
import numpy as np
from sklearn import metrics

# Define the categories
categories = ['angry', 'disgusted', 'fearful', 'happy', 'neutral', 'sad', 'surprised']

# Create the confusion matrix based on provided data
confusion_matrix = np.array([
    [40, 0, 0, 1, 0, 3, 2],  # Actual: Anger
    [1, 35, 2, 0, 2, 2, 0],  # Actual: Disgust
    [2, 2, 44, 0, 1, 0, 3],  # Actual: Fearful
    [0, 0, 0, 40, 1, 1, 0],  # Actual: Happy
    [2, 0, 0, 1, 32, 3, 2],  # Actual: Neutral
    [0, 0, 0, 0, 2, 36, 0],  # Actual: Sad
    [1, 0, 1, 0, 0, 1, 39],  # Actual: Surprise
])

# Display the confusion matrix
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=categories)
cm_display.plot(cmap=plt.cm.Blues)

# Show the plot
plt.show()

# Save the plot to a file
plt.savefig("confusion_matrix.png")
```
